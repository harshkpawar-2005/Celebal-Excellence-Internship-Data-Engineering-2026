## Data Cleaning

The dataset contains missing values and some columns require preprocessing before analysis. In this section, missing values are handled and price-related columns are cleaned.

```python
print("Duplicate Records:", df.duplicated().sum())
```

```python
df["final_price"] = (
    df["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
)

df["final_price"] = pd.to_numeric(
    df["final_price"],
    errors="coerce"
)

df["final_price"].head()
```

```python
df["final_price"].dtype
```

```python
df["discount"] = df["discount"].fillna(0)

df["seller_name"] = df["seller_name"].fillna("Unknown Seller")

df["seller_information"] = df["seller_information"].fillna("Not Available")
```

```python
df[["discount", "seller_name", "seller_information"]].isnull().sum()
```

## Feature Engineering

Additional columns are created to help analyze pricing and product popularity.

```python
df["price_difference"] = (
    df["initial_price"] -
    df["final_price"]
)

df[["initial_price", "final_price", "price_difference"]].head()
```

```python
df["popularity_score"] = (
    df["rating"] *
    df["ratings_count"]
)

df[["rating", "ratings_count", "popularity_score"]].head()
```

### Feature Engineering Output

The newly created columns are displayed below to verify the transformations.

```python
df[
    [
        "initial_price",
        "final_price",
        "price_difference",
        "rating",
        "ratings_count",
        "popularity_score"
    ]
].head()
```

## Data Analysis

After cleaning the dataset, exploratory analysis is performed to understand pricing patterns, product popularity, and category-level trends.

```python
top_categories = df["category"].value_counts()

top_categories.head(10)
```

```python
category_price = (
    df.groupby("category")["final_price"]
    .mean()
    .sort_values(ascending=False)
)

category_price.head(10)
```

```python
df[
    ["title", "rating", "ratings_count", "popularity_score"]
].sort_values(
    by="popularity_score",
    ascending=False
).head(10)
```

```python
df["final_price"].describe()
```

## Data Visualization

Visualizations are used to better understand product pricing and category distributions.

```python
plt.figure(figsize=(8,5))

plt.hist(df["final_price"], bins=30)

plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Number of Products")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["final_price"])

plt.title("Boxplot of Product Prices")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x=top_categories.head(10).index,
    y=top_categories.head(10).values
)

plt.xticks(rotation=45)

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="rating",
    y="final_price"
)

plt.title("Rating vs Final Price")
plt.xlabel("Rating")
plt.ylabel("Price")

plt.show()
```
